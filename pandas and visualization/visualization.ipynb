```python
#loading rice production across different countries from 1961 to 2021
df=pd.read_csv('rice production across different countries from 1961 to 2021.csv')
df.columns
```

```python
#information about the data
df.describe()
```

```python
df.info()
```

```python
df2=df[['Area','Year','Value']]
#filtering for 2021
df2=df2.loc[df2['Year']==2021]
df2=df2.loc[df2['Area'].isin(['Angola', 'Brunei Darussalam','Belize'])]
df2
```

```python
x=df2['Area']
y=df2['Value']
plt.bar(x,y)
```

```python
#Analysing production in nigeria since 1960-2021
df3=df.loc[df['Area']=='Nigeria']
df3
```

```python
#plotting year against value
#Line Graph
x=df3['Year']
y=df3['Value']
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('plotting year against value')
plt.plot(x,y)
```

```python
#plotting year against value
#Bar graph
x=df3['Year']
y=df3['Value']
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('plotting year against value')
plt.bar(x,y)
```
