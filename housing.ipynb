```python
#Importing data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the data
df=pd.read_csv('housing.csv')
df
```

```python
#data description
df.describe()
```

```python
#info
df.info()
```

```python
#fill empty rows with the median value 
median = df['total_bedrooms'].median()
df['total_bedrooms'].fillna(median , inplace = True)
```

```python
#data visualization
x=df['housing_median_age']
y=df['median_house_value']
df.hist(figsize=(10,10))
plt.show()
```

```python
#Building the model
#defining the input x and the prediction y
y=df['median_house_value']
x = df.drop(['median_house_value', 'households','population','ocean_proximity'], axis = 1)
x
```

```python
#Training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=2)
```

```python
#train the data 
#call the fit method on the model and pass the training data to it 
model.fit(x_train, y_train)
```

```python
#test and score the model against the test data
model.score(x_test, y_test)
```
